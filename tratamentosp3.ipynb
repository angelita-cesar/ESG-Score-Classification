```python
#bliblioteca para manipulação de dados e leitura de arquivos CSV
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\mikae\\OneDrive\\Área de Trabalho\\CESAR\\Projetos 3\\datakaggle.csv")
```

```python
#impressao de valores únicos para cada coluna para possibilitar analises posteriores
for coluna in df.columns:
    print(f"Valores únicos na coluna '{coluna}':")
    print(df[coluna].unique())
    print("-" * 30) 
```

```python
#drop D colunas de apenas texto e labels que não são uteis ou são repetidas.
colunas_para_dropar = ['ticker', 'currency', 'exchange', 'logo', 'weburl', 'cik','last_processing_date']
df = df.drop(columns=colunas_para_dropar)
```

```python
#utilizando df.shape para verificar o número de linhas e colunas do DataFrame após limpeza.
print("Formato do DataFrame após limpeza:")
print(df.shape)
```

```python
# verificação de valores ausentes para tratamento posterior
print(df.isna().sum())

for col in df.columns:
    if df[col].isna().sum() > 0:
        print(f"Coluna '{col}' tem {df[col].isna().sum()} valores ausentes.")
```

```python
# 1. Filtrar as 13 linhas onde a coluna 'industry' é NULA
df_nulos = df[df["industry"].isna()]

# 2. Filtrar o restante das linhas onde a coluna 'industry' NÃO é nula
df_limpo = df[df["industry"].notna()]

# 3. Salvar as linhas nulas em um novo CSV
df_nulos.to_csv(r"C:\\Users\\mikae\\OneDrive\\Área de Trabalho\\CESAR\\Projetos 3\\dados_com_nulos.csv", index=False)

# 4. Salvar a base limpa em outro CSV
df_limpo.to_csv(r"C:\\Users\\mikae\\OneDrive\\Área de Trabalho\\CESAR\\Projetos 3\\base_tratada.csv", index=False)

#Validação (opcional, mas recomendada para garantir que os números batem)
print(# 13
    f"Linhas nulas exportadas: {len(df_nulos)}"
)  
print(f"Linhas limpas exportadas: {len(df_limpo)}")
```
